```python
from datasets import load_dataset
from datasets import Audio

minds = load_dataset("PolyAI/minds14", name="en-AU", split="train")
minds = minds.cast_column("audio", Audio(sampling_rate=16_000))
```

```python
from transformers import pipeline

classifier = pipeline(
    "audio-classification",
    model="anton-l/xtreme_s_xlsr_300m_minds14",
)
```

```python
example = minds[0]
```

```python
classifier(example["audio"]["array"])
```

```python
id2label = minds.features["intent_class"].int2str
id2label(example["intent_class"])
```

```python
from transformers import pipeline

pipe = pipeline("text-to-speech", model="suno/bark-small")
```

```python
text = "Ladybugs have had important roles in culture and religion. "
output = pipe(text)
```

```python
from IPython.display import Audio

Audio(output["audio"].squeeze(), rate=output["sampling_rate"])
```

```python
es_text = "Contrariamente a la creencia popular, el número de manchas en los élitros de una mariquita no corresponde a su edad, ni en años ni en meses. "
output = pipe(es_text)
Audio(output["audio"].squeeze(), rate=output["sampling_rate"])
```

```python
song = "♪ In the jungle, the mighty jungle, the ladybug was seen. ♪ "
output = pipe(song)
Audio(output["audio"].squeeze(), rate=output["sampling_rate"])
```
